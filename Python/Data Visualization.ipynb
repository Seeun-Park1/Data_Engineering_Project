### 분산 과 표준편차

```python
# 분산 variance
# 표준편차 standard_deviation

st1Mean = np.mean(st1)
st1Len = len(st1)
diffSum = 0
for eachitem in st1:
    diffSum = diffSum + (st1Mean-eachitem) ** 2
```

```python
stdValue = np.sqrt(diffSum / (st1Len-1))
stdValue
```

# 2. matplotlib 활용 시각화 고급

### (1) 서브플롯 그리기

```python
import pandas as pd

st0001LocalUrl = "../dataset/ST0001DATA.csv"
st0002LocalUrl = "../dataset/ST0002DATA.csv"
st0001GitUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/st0001data.csv"
st0002GitUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/st0002data.csv"

groupDataS01 = pd.read_csv(st0001GitUrl)
groupDataS02 = pd.read_csv(st0002GitUrl)

# 필요 라이브러리 정의
import matplotlib.pyplot as plt
# 팝업 창 활용하여 차트 시연
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME
y2=groupDataS02.MEAN_VOLUME

fig = plt.figure(figsize=(14,4))
# 첫번째 차트
ax1 = fig.add_subplot(121)
ax1.plot(x,y, 'bo-',lw=1.5, label = 'ST0001')
ax1.grid(True)
ax1.legend(loc=0)
ax1.set_xlabel('ax1_index')
ax1.set_ylabel('ax1_value')
ax1.set_title('ax1_title')
# 두번째 차트
ax2 = fig.add_subplot(122)
ax2.plot(x,y2, 'go-', lw=1.5, label='ST0002')
ax2.grid(True)
ax2.legend(loc=0)
ax2.set_xlabel('ax2_index')
ax2.set_ylabel('ax2_value')
ax2.set_title('ax2_title')
```

### (2) 이중축 그리기

```python
# import matplotlib
# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 14}
```

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 팝업 창 활용하여 차트 시연
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME
y2=groupDataS02.MEAN_VOLUME

# 이중축
fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(1,1,1)
#figure, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(x,y, 'bo-',lw=1.5, label = 'ST0001')
ax1.grid(True)
ax1.legend(loc=0)
ax1.set_ylabel('ax1_value')
ax1.set_xlabel('ax1_index')
plt.title('subplot')
ax2 = ax1.twinx()
ax2.plot(x,y2, 'go-', lw=1.5, label='ST0002')
ax2.legend(loc=4)
ax2.set_ylabel('ax2_value')
```

# 3. Seaborn

```python
import seaborn as sns
import pandas as pd
%matplotlib inline

customerLocalUrl = "../dataset/customerdata.csv"
customerGitUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/customerdata.csv"
csData = pd.read_csv(customerGitUrl)
csData.head(2)
```

### (1) 카운트 차트(count plot)

```python
sns.countplot(data = csData, x="EMI", hue="CUSTTYPE")
```

### (2) 히스토그램 (dist plot)

```python
figure = plt.figure(figsize=(18, 6))

# 히스토그램 그래프 (kde=False)
ax1 = plt.subplot(1, 2, 1)
sns.histplot(csData.AVGPRICE, kde=False, ax=ax1)

# KDE 그래프 (hist=False)
ax2 = plt.subplot(1, 2, 2)
sns.kdeplot(csData.AVGPRICE, ax=ax2)
```

### (3) 바 차트 (bar plot)

```python
sns.barplot(x="EMI", y="AVGPRICE", data = csData,
         hue = "CUSTTYPE", ci = "sd")
```

### (4) 박스 차트 (box plot)

```python
sns.boxplot(data=csData, x="EMI", y="AVGPRICE")
```

### (5) 선 차트 (line plot)
